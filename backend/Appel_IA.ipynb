# Launch inference

```python
API_ENDPOINT = URL + "inferences"
print (API_ENDPOINT)
filename = 'filetoIA.csv'  
data = {
 "network_name": "Network.002"
}
files = {
    'history': (filename, open(filename, 'rb')),
}
r = requests.post(url = API_ENDPOINT, data = data, files = files,  headers = keyId)
print(r.status_code)
print(r.text)
token = r.json()['token']

#########################################################################
# TASK
#########################################################################
task_status = "CREATED"
API_TASK = URL + "tasks/" + token
outcome_status = list()
outcome_status.append("DONE")
outcome_status.append("ERROR")
print(datetime.now())
task_status = "";
while task_status not in outcome_status:
  r = requests.get(url = API_TASK,  headers = keyId)
  if not(task_status == r.json()["status"]):
      task_status = r.json()["status"]
      print("status is " + task_status + ", sleeping...")
  time.sleep(1)
print(r.json())
print(datetime.now())
```

# Get Inference Result

```python
API_ENDPOINT = URL + "/tasks/" + token
r = requests.get(url = API_ENDPOINT,  headers = keyId)
API_ENDPOINT = URL + "inferences/"+r.json()['resourceId']
r2=requests.get(url = API_ENDPOINT,  headers = keyId)
print(json.dumps(r2.json(), sort_keys=True, indent=4))
```
